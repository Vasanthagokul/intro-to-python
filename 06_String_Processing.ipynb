## 5. Data Validation ğŸ›¡ï¸

Check if user input is valid and clean!

```python
def validate_email(email):
    """Simple email validation"""
    email = email.strip().lower()
    
    # Basic checks
    if '@' not in email:
        return False, "Missing @ symbol"
    
    if email.count('@') != 1:
        return False, "Multiple @ symbols"
    
    if email.startswith('@') or email.endswith('@'):
        return False, "@ symbol at wrong position"
    
    if '.' not in email.split('@')[1]:
        return False, "Missing domain extension"
    
    return True, "Valid email"

# Test email validation
test_emails = [
    "student@college.edu",
    "invalid-email",
    "test@@example.com",
    "@invalid.com",
    "valid@domain.org",
    "no-dot@domain"
]

print("Email validation results:")
for email in test_emails:
    is_valid, message = validate_email(email)
    status = "âœ“" if is_valid else "âœ—"
    print(f"{status} {email:<25} - {message}")
```

### Phone Number Validation

```python
def validate_indian_phone(phone):
    """Validate Indian phone numbers"""
    # Clean the phone number
    clean_phone = phone.replace(' ', '').replace('-', '').replace('+91', '')
    
    # Check if only digits
    if not clean_phone.isdigit():
        return False, "Contains non-digit characters"
    
    # Check length
    if len(clean_phone) != 10:
        return False, f"Should be 10 digits, got {len(clean_phone)}"
    
    # Check if starts with valid digit (6,7,8,9)
    if clean_phone[0] not in '6789':
        return False, "Should start with 6, 7, 8, or 9"
    
    return True, f"Valid phone: {clean_phone}"

# Test phone validation
test_phones = [
    "9876543210",
    "+91-9876543210",
    "98765 43210",
    "1234567890",  # Invalid start
    "98765432",    # Too short
    "98765432101", # Too long
    "987654321a"   # Contains letter
]

print("\nPhone validation results:")
for phone in test_phones:
    is_valid, message = validate_indian_phone(phone)
    status = "âœ“" if is_valid else "âœ—"
    print(f"{status} {phone:<20} - {message}")
```

### Roll Number Validation

```python
def validate_roll_number(roll):
    """Validate roll number format: CS2023001"""
    roll = roll.upper().strip()
    
    # Check length
    if len(roll) != 9:
        return False, f"Should be 9 characters, got {len(roll)}"
    
    # Check department code
    dept_code = roll[:2]
    valid_depts = ['CS', 'EC', 'ME', 'CE', 'EE']
    if dept_code not in valid_depts:
        return False, f"Invalid department code: {dept_code}"
    
    # Check year
    year_part = roll[2:6]
    if not year_part.isdigit():
        return False, "Year part should be digits"
    
    # Check sequence number
    seq_part = roll[6:]
    if not seq_part.isdigit():
        return False, "Sequence part should be digits"
    
    return True, f"Valid roll number: {roll}"

# Test roll number validation
test_rolls = [
    "CS2023001",
    "cs2023001",  # lowercase
    "EC2024050",
    "XY2023001",  # invalid dept
    "CS20230",    # too short
    "CS202300A",  # contains letter
]

print("\nRoll number validation results:")
for roll in test_rolls:
    is_valid, message = validate_roll_number(roll)
    status = "âœ“" if is_valid else "âœ—"
    print(f"{status} {roll:<15} - {message}")
```

```python
# Exercise 1: Name Formatter
def format_name(name):
    # Clean and normalize
    clean_name = name.strip()
    
    # Replace separators with spaces
    clean_name = clean_name.replace('-', ' ').replace('_', ' ').replace('.', ' ')
    
    # Split into parts and remove empty strings
    parts = [part.strip() for part in clean_name.split() if part.strip()]
    
    # Convert to title case
    parts = [part.title() for part in parts]
    
    if len(parts) == 1:
        return parts[0]
    elif len(parts) == 2:
        return f"{parts[1]}, {parts[0]}"
    else:
        # Multiple names: Last, First Middle...
        last = parts[-1]
        first_middle = ' '.join(parts[:-1])
        return f"{last}, {first_middle}"

print("Exercise 1: Name Formatter")
messy_names = [
    "  ALICE   MARIE   JOHNSON  ",
    "bob-smith",
    "carol_davis_brown",
    "diana.wilson.jr",
    "EDWARD VAN DER BERG"
]

for name in messy_names:
    print(f"'{name}' -> '{format_name(name)}'")

print("\n" + "="*60 + "\n")

# Exercise 2: Text Statistics Analyzer
def analyze_text(text):
    # Clean text
    clean_text = text.strip()
    
    # Count paragraphs (by double newlines)
    paragraphs = [p.strip() for p in clean_text.split('\n\n') if p.strip()]
    paragraph_count = len(paragraphs)
    
    # Count sentences (basic - by periods, exclamation, question marks)
    sentence_endings = clean_text.count('.') + clean_text.count('!') + clean_text.count('?')
    
    # Count words
    words = clean_text.lower().replace('.', '').replace(',', '').replace('!', '').replace('?', '').split()
    word_count = len(words)
    
    # Word frequency
    word_freq = {}
    for word in words:
        word_freq[word] = word_freq.get(word, 0) + 1
    
    # Most frequent word
    most_frequent = max(word_freq, key=word_freq.get) if word_freq else ""
    
    # Python count
    python_count = clean_text.lower().count('python')
    
    # Average words per sentence
    avg_words_per_sentence = word_count / sentence_endings if sentence_endings > 0 else 0
    
    # Generate report
    report = f"""
TEXT ANALYSIS REPORT
====================
Total Paragraphs: {paragraph_count}
Total Sentences: {sentence_endings}
Total Words: {word_count}
Average Words per Sentence: {avg_words_per_sentence:.1f}

Most Frequent Word: '{most_frequent}' ({word_freq.get(most_frequent, 0)} times)
Mentions of 'Python': {python_count}

Top 5 Most Frequent Words:
"""
    
    # Top 5 words
    sorted_words = sorted(word_freq.items(), key=lambda x: x[1], reverse=True)[:5]
    for word, count in sorted_words:
        report += f"  {word}: {count}\n"
    
    return report

sample_text = """
Python is an amazing programming language. It's easy to learn and powerful to use.
Python developers love its simplicity and readability. Many companies use Python
for web development, data science, artificial intelligence, and automation.
Learning Python opens many career opportunities!
"""

print("Exercise 2: Text Analysis")
print(analyze_text(sample_text))

print("\n" + "="*60 + "\n")

# Exercise 3: Password Strength Checker
def check_password_strength(password):
    score = 0
    feedback = []
    
    # Length check
    if len(password) >= 8:
        score += 20
    else:
        feedback.append("Add more characters (minimum 8)")
    
    # Uppercase check
    if any(c.isupper() for c in password):
        score += 20
    else:
        feedback.append("Add uppercase letters")
    
    # Lowercase check
    if any(c.islower() for c in password):
        score += 20
    else:
        feedback.append("Add lowercase letters")
    
    # Digit check
    if any(c.isdigit() for c in password):
        score += 20
    else:
        feedback.append("Add numbers")
    
    # Special character check
    special_chars = "!@#$%^&*()_+-=[]{}|;':,.<>?"
    if any(c in special_chars for c in password):
        score += 20
    else:
        feedback.append("Add special characters")
    
    # Common weak patterns
    weak_patterns = ['password', '123456', 'qwerty', 'abc']
    if any(pattern in password.lower() for pattern in weak_patterns):
        score -= 30
        feedback.append("Avoid common patterns")
    
    # Length bonus
    if len(password) >= 12:
        score += 10
    
    score = max(0, min(100, score))  # Keep score between 0-100
    
    if score >= 80:
        strength = "Strong"
    elif score >= 60:
        strength = "Medium"
    else:
        strength = "Weak"
    
    feedback_text = "; ".join(feedback) if feedback else "Excellent password!"
    
    return score, f"{strength} - {feedback_text}"

print("Exercise 3: Password Strength")
test_passwords = [
    "password123",
    "MySecure@Pass123",
    "weak",
    "Strong@Password2024!",
    "NoNumbers@",
    "nouppercasehere123@"
]

for pwd in test_passwords:
    score, feedback = check_password_strength(pwd)
    print(f"'{pwd:<20}' -> {score:3d}/100 - {feedback}")

print("\n" + "="*60 + "\n")

# Exercise 4: CSV Data Cleaner
def clean_student_data(data):
    # Department mapping
    dept_map = {
        "computer science": "CS",
        "mechanical engineering": "ME",
        "electrical & electronics": "EE",
        "information technology": "IT"
    }
    
    lines = [line.strip() for line in data.strip().split('\n') if line.strip()]
    cleaned_records = []
    
    print("Clean CSV Data:")
    print("Last, First,Dept,CGPA,Email,Phone")
    print("-" * 50)
    
    for line in lines:
        # Split by comma
        fields = [field.strip() for field in line.split(',')]
        
        # Clean name
        name_parts = fields[0].strip().title().split()
        if len(name_parts) >= 2:
            formatted_name = f"{name_parts[-1]}, {' '.join(name_parts[:-1])}"
        else:
            formatted_name = name_parts[0] if name_parts else ""
        
        # Clean department
        dept_full = fields[1].lower().strip()
        dept_code = dept_map.get(dept_full, dept_full.upper()[:2])
        
        # Clean CGPA
        cgpa = fields[2].strip()
        
        # Clean email
        email = fields[3].strip().lower()
        
        # Clean phone
        phone = fields[4].replace('+91', '').replace('-', '').replace(' ', '').strip()
        
        print(f"{formatted_name},{dept_code},{cgpa},{email},{phone}")
        
        cleaned_records.append({
            'name': formatted_name,
            'department': dept_code,
            'cgpa': cgpa,
            'email': email,
            'phone': phone
        })
    
    return cleaned_records

print("Exercise 4: Data Cleaning")
messy_data = """
alice JOHNSON,  Computer Science , 8.5 ,alice@college.edu,  9876543210
BOB smith,mechanical engineering,7.8,bob.smith@email.com,+91-8765432109  
Carol  DAVIS,   electrical & electronics,9.2,  carol.d@college.edu ,8654321098
diana Wilson,Information Technology,8.9, diana@it.college.edu,+91 7543210987
"""

clean_student_data(messy_data)

print("\n" + "="*60 + "\n")

# Exercise 5: Academic Report Generator
from datetime import datetime

def calculate_letter_grade(percentage):
    if percentage >= 90: return "A+"
    elif percentage >= 85: return "A"
    elif percentage >= 80: return "B+"
    elif percentage >= 75: return "B"
    elif percentage >= 70: return "C+"
    elif percentage >= 60: return "C"
    else: return "F"

def calculate_cgpa(subjects):
    total = sum(subjects.values())
    average = total / len(subjects)
    return (average / 100) * 4.0  # Convert to 4.0 scale

def generate_report_card(student, rank=None):
    name = student['name']
    roll = student['roll']
    subjects = student['subjects']
    
    cgpa = calculate_cgpa(subjects)
    average = sum(subjects.values()) / len(subjects)
    letter_grade = calculate_letter_grade(average)
    
    current_date = datetime.now().strftime("%B %d, %Y")
    
    report = f"""
â•”â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•—
â•‘                        STUDENT REPORT CARD                   â•‘
â• â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•£
â•‘ Name: {name:<30} Roll: {roll:<15} â•‘
â•‘ Date: {current_date:<49} â•‘
â• â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•£
â•‘                          GRADES                              â•‘
â• â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•£"""
    
    for subject, marks in subjects.items():
        grade = calculate_letter_grade(marks)
        report += f"\nâ•‘ {subject:<20} {marks:>3d}%     Grade: {grade:<15} â•‘"
    
    report += f"""
â• â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•£
â•‘ Average: {average:5.1f}%     CGPA: {cgpa:.2f}/4.0     Overall: {letter_grade:<8} â•‘"""
    
    if rank:
        report += f"\nâ•‘ Class Rank: {rank:<44} â•‘"
    
    report += """
â•šâ•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•"""
    
    return report

def generate_class_summary(students):
    # Calculate rankings
    student_averages = []
    for student in students:
        avg = sum(student['subjects'].values()) / len(student['subjects'])
        student_averages.append((student, avg))
    
    # Sort by average (descending)
    student_averages.sort(key=lambda x: x[1], reverse=True)
    
    current_date = datetime.now().strftime("%B %d, %Y")
    
    summary = f"""
â•”â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•—
â•‘                      CLASS SUMMARY REPORT                    â•‘
â•‘                         {current_date:<30}               â•‘
â• â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•£
â•‘ Rank â”‚ Name              â”‚ Roll No    â”‚ Average â”‚ Grade      â•‘
â• â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•£"""
    
    for i, (student, avg) in enumerate(student_averages, 1):
        name = student['name']
        roll = student['roll']
        grade = calculate_letter_grade(avg)
        summary += f"\nâ•‘ {i:2d}   â”‚ {name:<17} â”‚ {roll:<10} â”‚ {avg:5.1f}%  â”‚ {grade:<10} â•‘"
    
    # Class statistics
    all_averages = [avg for _, avg in student_averages]
    class_avg = sum(all_averages) / len(all_averages)
    highest = max(all_averages)
    lowest = min(all_averages)
    
    summary += f"""
â• â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•£
â•‘                      CLASS STATISTICS                        â•‘
â• â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•£
â•‘ Class Average: {class_avg:5.1f}%                                    â•‘
â•‘ Highest Score: {highest:5.1f}%                                    â•‘
â•‘ Lowest Score:  {lowest:5.1f}%                                     â•‘
â•‘ Total Students: {len(students):<2d}                                      â•‘
â•šâ•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•"""
    
    return summary

print("Exercise 5: Academic Reports")
student_data = [
    {"name": "Alice Johnson", "roll": "CS2023001", "subjects": {"Math": 89, "Physics": 92, "Chemistry": 85, "CS": 95}},
    {"name": "Bob Smith", "roll": "ME2023002", "subjects": {"Math": 76, "Physics": 81, "Chemistry": 78, "Engineering": 84}},
    {"name": "Carol Davis", "roll": "EE2023003", "subjects": {"Math": 94, "Physics": 97, "Chemistry": 91, "Electronics": 89}}
]

# Generate individual reports with ranks
student_averages = [(student, sum(student['subjects'].values()) / len(student['subjects'])) for student in student_data]
student_averages.sort(key=lambda x: x[1], reverse=True)

for i, (student, avg) in enumerate(student_averages, 1):
    print(generate_report_card(student, rank=f"#{i} out of {len(student_data)} students"))
    print("\n")

# Generate class summary
print(generate_class_summary(student_data))
```
