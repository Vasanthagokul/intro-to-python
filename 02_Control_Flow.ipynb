## 1. Conditional Statements (if/elif/else)

Make decisions in your code based on conditions.

````python
# Simple if statement
temperature = 35

if temperature > 30:
    print("It's hot outside!")
    print("Drink plenty of water")
````

````python
# if-else statement
age = 17

if age >= 18:
    print("You can vote")
else:
    print("You cannot vote yet")
    print(f"Wait {18 - age} more year(s)")
````

````python
# if-elif-else statement
marks = 75

if marks >= 90:
    grade = "A+"
elif marks >= 80:
    grade = "A"
elif marks >= 70:
    grade = "B"
elif marks >= 60:
    grade = "C"
else:
    grade = "F"

print(f"Marks: {marks}, Grade: {grade}")
````

### Nested if Statements

````python
age = 20
has_license = True

if age >= 18:
    if has_license:
        print("You can drive")
    else:
        print("Get a license first")
else:
    print("You're too young to drive")
````

### Multiple Conditions with Logical Operators

````python
# AND operator - both conditions must be True
age = 20
has_license = True

if age >= 18 and has_license:
    print("You can drive")
else:
    print("You cannot drive")

# OR operator - at least one condition must be True
day = "Saturday"

if day == "Saturday" or day == "Sunday":
    print("It's a weekend!")
else:
    print("It's a weekday")
````

## 2. For Loops

Iterate over a sequence (list, string, range, etc.)

````python
# Loop through a range of numbers
print("Counting from 1 to 5:")
for i in range(1, 6):
    print(i)
````

````python
# Loop through a list
fruits = ["apple", "banana", "mango", "orange"]

print("Fruits available:")
for fruit in fruits:
    print(f"- {fruit}")
````

````python
# Loop through a string
word = "PYTHON"

for letter in word:
    print(letter)
````

### Range Function

The `range()` function generates a sequence of numbers.

````python
# range(stop) - starts from 0
print("range(5):")
for i in range(5):
    print(i, end=" ")  # 0 1 2 3 4

print("\n\nrange(start, stop):")
for i in range(2, 8):
    print(i, end=" ")  # 2 3 4 5 6 7

print("\n\nrange(start, stop, step):")
for i in range(0, 11, 2):
    print(i, end=" ")  # 0 2 4 6 8 10
````

### Enumerate Function

Get both index and value when looping.

````python
subjects = ["Math", "Physics", "Chemistry", "CS"]

for index, subject in enumerate(subjects, start=1):
    print(f"{index}. {subject}")
````

## 3. While Loops

Repeat code while a condition is True.

````python
# Basic while loop
count = 1

while count <= 5:
    print(f"Count: {count}")
    count += 1  # Same as: count = count + 1

print("Loop finished!")
````

````python
# User input validation
password = ""

while password != "python123":
    password = input("Enter password: ")
    if password != "python123":
        print("Wrong password! Try again.")

print("Access granted!")
````

## 4. Break and Continue

Control loop execution with `break` and `continue`.

### Break - Exit the loop early

````python
# Search for a number
numbers = [10, 25, 30, 45, 50, 60]
search = 45

for num in numbers:
    if num == search:
        print(f"Found {search}!")
        break
    print(f"Checking {num}...")
else:
    # This runs only if loop completes without break
    print(f"{search} not found")
````

### Continue - Skip current iteration

````python
# Print only odd numbers
for i in range(1, 11):
    if i % 2 == 0:  # If even, skip
        continue
    print(i, end=" ")  # Only odd numbers printed
````

### Nested Loops

````python
# Multiplication table
for i in range(1, 4):
    for j in range(1, 4):
        print(f"{i} Ã— {j} = {i*j}")
    print()  # Empty line after each table
````
